The Project is about analysing the Data of the Six Differnt Countires. It involves answering Four main questions such as 

1. Has life expectancy increased over time in the six nations?
2. Has GDP increased over time in the six nations?
3. Is there a correlation between GDP and life expectancy of a country?
4. What is the distribution of that life expectancy?

```python
#importing all the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
# Read the data

data = pd.read_csv('all_data.csv')
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
# Has life expectancy increased over time in the six nations?


plt.figure(figsize = (8,8))
sns.scatterplot(data = data, y = "Life expectancy at birth (years)", x="Year", hue ="Country")
plt.show()
plt.clf()
```

```python
# Has GDP increased over time in the six nations?

plt.figure(figsize = (10,10))
sns.lmplot(x = 'Year', y = 'GDP',  hue = 'Country', data = data, lowess =True)
plt.show()
plt.clf()
```

```python
#Is there a correlation between GDP and life expectancy of a country?
log_gdp = np.log(data.GDP)

from scipy.stats import pearsonr
corr, p = pearsonr(data['GDP'],data['Life expectancy at birth (years)'])
print(corr)   ##return between -1 to +1"

```

```python
#What is the average life expectancy in these nations?

plt.figure(figsize = (8,8))
ax= plt.subplot()
sns.boxplot(data = data, y = "Life expectancy at birth (years)", x="Country", orient = "v")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
#What is the distribution of that life expectancy?

plt.figure(figsize = (8,8))
sns.distplot(data['Life expectancy at birth (years)'])
plt.show()

sns.violinplot(x=data['Life expectancy at birth (years)'])
plt.show()
```
